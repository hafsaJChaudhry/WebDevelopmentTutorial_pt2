## How to Run App.py without using Flask Run Command

1. **Create Virtual Environment**: (shortened as "virtualenv") separate our new project's Python dependencies from our other projects and from the Python libraries our operating system uses. If you don't use a virtualenv, there's a good chance you might break part of your OS.

On the terminal that you're running your code, put in these commands:

> python3 -m venv venv

> source venv/bin/activate

> (venv)(base) will show, now write pip3 install flask

```python
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "Hello, World"

#if __name__= "__main__" will be the case if we run app.py
# host='0.0.0.0' == host='localhost' which allows the code to run locally on your computer
# debug=True allows you to make changes
if __name__ == "__main__":
    app.run(host='0.0.0.0', debug=True)
```
